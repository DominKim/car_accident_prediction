```python
data_dict = {"index" : [], "value" : [], "file_name" : []}
def file_name():
    lst = os.listdir("./data")
    name = []
    for num in range(1, 35):
        for check in lst:
            if num == int(check.split(".")[0]):
                name.append(check)
    return name
```
